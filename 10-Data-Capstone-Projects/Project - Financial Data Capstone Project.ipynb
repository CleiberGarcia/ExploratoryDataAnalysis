# Import Python Modules and Setup

```python
from pandas_datareader import data
import yfinance as yf
yf.pdr_override() 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

## Loading Data for BAC, C, GS, JPM, MS and WFC

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2022, 12, 31)
```

```python
# Bank of America
symbol = 'BAC'
start_date = start
end_date = end
BAC = data.get_data_yahoo(symbol, start_date, end_date)
BAC.head()
```

```python
# CitiGroup
symbol = 'C'
start_date = start
end_date = end
C = data.get_data_yahoo(symbol, start_date, end_date)
C.head()
```

```python
# Goldman Sachs
symbol = 'GS'
start_date = start
end_date = end
GS = data.get_data_yahoo(symbol, start_date, end_date)
GS.head()
```

```python
# JPMorgan Chase
symbol = 'JPM'
start_date = start
end_date = end
JPM = data.get_data_yahoo(symbol, start_date, end_date)
JPM.head()
```

```python
# Morgan Stanley
symbol = 'MS'
start_date = start
end_date = end
MS = data.get_data_yahoo(symbol, start_date, end_date)
MS.head()
```

```python
# Wells Fargo
symbol = 'WFC'
start_date = start
end_date = end
WFC = data.get_data_yahoo(symbol, start_date, end_date)
WFC.head()
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], keys=tickers, axis=1)
bank_stocks.head()
```

** Set the column name levels (this is filled out for you):**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

** Check the head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
    
returns.head()
```

** Create a pairplot using seaborn of the returns dataframe

```python
sns.pairplot(returns)
```

## Moving Averages
