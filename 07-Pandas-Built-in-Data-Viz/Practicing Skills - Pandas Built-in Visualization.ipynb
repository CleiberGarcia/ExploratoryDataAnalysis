# 1. Importing Python Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2. Loading Working Data Set

```python
df3 = pd.read_csv('df3')
df3.head()
```

```python
df3.info()
```

# 3. Practice Plots

## 3.1 Scatter Plot

```python
plt.figure(figsize=(12,3))
df3.plot.scatter(x='a', y='b', c='red', s=50, figsize=(10,3))
```

## 3.2 Histogram

```python
plt.style.use('ggplot')
df3['a'].hist()
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=50)
```

## 3.3 BoxPlot

```python
df3[['a','b']].plot.box()
```

## 3.4 kde

```python
df3['d'].plot.kde()
```

```python
df3['d'].plot.density(lw=5, ls='--')
```

## 3.5 Area Plot

```python
df3[:30].plot.area(alpha=0.4)
```

```python
df3[:30].plot.area(alpha=0.4).legend(bbox_to_anchor=(1.0, 1.0))
```
