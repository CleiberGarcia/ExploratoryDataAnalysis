## Importing Python modules

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Working Data

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

## Practice 1

* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Graphic 1')

plt.plot(x,y)
```

## Practice 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**
** Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Creates an empty canvas
fig = plt.figure()

# Add axes to figure
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('X ext')
ax1.set_ylabel('Y ext')
ax1.set_title('Graphic 2 - ext')
ax1.plot(x, y, 'b')

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.set_xlabel('X int')
ax2.set_ylabel('Y int')
ax2.set_title('Graphic 2 - int')
ax2.plot(x, y, 'g')
```

## Practice 3

** Create a plot by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**
** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Manually adding axes with add_axes is can be useful for adding insets to figures
fig = plt.figure()

# Add axes to figure
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('X ext')
ax1.set_ylabel('Y ext')
ax1.set_title('Graphic 3')
ax1.plot(x,z)

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
# set axis range
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
ax2.set_xlabel('X int')
ax2.set_ylabel('Y int')
ax2.set_title('zoom')
ax2.plot(x,y)

#fig
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

axes[0].plot(x,y)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Graphic X by Y')

axes[1].plot(x,z)
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_title('Graphic X by Z')

plt.tight_layout()
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

axes[0].plot(x,y,color="blue", lw=5, ls='-.')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Graphic X by Y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_title('Graphic X by Z')

plt.tight_layout()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,3))
plt.tight_layout()

axes[0].plot(x,y,color="blue", lw=5, ls='-.')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Graphic X by Y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].set_title('Graphic X by Z')

plt.tight_layout()
```
